```python
 # importing libraries
import pandas as pd
import numpy as np
```

```python
# turning the dictionary into a DataFrame
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(exam_data)
df.head()
```

```python
# 2 - Delete rows with Nan values.

df = df.dropna() # drop the rows with NAN values
df
```

```python
# 3 - Extract the 'name' and 'score' columns from the DataFrame.
df.iloc[:, 0:2]
```

```python
#4-Write a Pandas program to append a new row 'k' to the DataFrame with these values
dict2 = {"name":["Suresh"],"score":[15.5],"attempts":[1],"qualify":["yes"]}

dict2_dataFrame = pd.DataFrame(dict2)

new_data = pd.concat([df,dict2_dataFrame])

data = dict(new_data)

df3 = pd.DataFrame(data)
df3
```

```python
# update the concatenated dataframes

def update(df3):
    x=len(df3)
    liste = []
    for i in range(x):
        liste.append(i)
    df3 = df3.set_index([pd.Index(liste)])
    return df3

df3 = update(df3)
df3
```

```python
#5 - Write a Pandas program to delete the 'attempts' column from the DataFrame.

df3.drop("attempts", axis=1)
```

```python
#6 - Add a new column "Success": if the score is higher than 10 we will have 1, else we will have 0.
df3["Success"] = 1
for i in df3.index:
    for j in range(3):
        if(df3.loc[i,"score"] < 10):
            df3.loc[i,"Success"] = 0
df3
```

```python
#7 - After executing the final DataFrame, export it into a CSV file named "my_data".

df3.to_csv('my_data.csv')
```
