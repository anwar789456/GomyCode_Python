```python
import requests
import string
from bs4 import BeautifulSoup
```

```python
# funtion that gets the user's input and returns URL
def Get_user_input():
    user_request = input("enter Search :")
    user_input = string.capwords(user_request)
    lists = user_input.split()
    word = "_".join(lists)
    url = "https://en.wikipedia.org/wiki/" + word
    return  url
url = Get_user_input()
```

```python
# function that takes a URL and get content from the page
def Get_Parse_Content(url):
    url_ = requests.get(url)
    soup = BeautifulSoup(url_.content, 'html.parser')
    details = soup('table',{'class':'infobox'})
    for i in details:
        h = i.find_all("tr")
        for j in h:
            heading = j.find_all('th')
            detail = j.find_all('td')
            if heading is not None and detail is not None:
                for x,y in zip(heading,detail):
                    print("{} :: {}".format(x.text,y.text))
                    print("----------------------")
    for i in range(1,3):
        print(soup('p')[i].text)
Get_Parse_Content(url)
```

```python
def Extract_Title(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    title_element = soup.find('h1')
    title = title_element.text.strip()
    return title

title = Extract_Title(url)
print(title)
```

```python
def extract_article_content_and_map(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    headings = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
    paragraphs = soup.find_all('p')
    dicte = {}
    for heading in headings:
        text = heading.text.strip()
        next_paragraph = heading.find_next_sibling('p')
        if next_paragraph:
            paragraph_text = next_paragraph.text.strip()
            dicte[text] = paragraph_text
    for paragraph in paragraphs:
        paragraph_text = paragraph.text.strip()
        if paragraph_text not in dicte.values():
            dicte[''] = paragraph_text
    return dicte
content = extract_article_content_and_map(url)

for heading, paragraph in content.items():
    print(f"{heading}: {paragraph}")
```

```python
def Get_Links(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    page_tags = soup.find_all('a')
    links = []
    for anchor_tag in page_tags:
        if 'href' in anchor_tag.attrs:
            href = anchor_tag['href']
            if href.startswith('/wiki/') and ':' not in href:
                links.append(href)
    return links
links = Get_Links(url)
for link in links:
    print(link)
```

```python
def wrapped_funtion():
    url = Get_user_input()
    print("what do you want?\n 1- see the content from Wikipedia\n 2- extract article title\n 3-Extract article text for each paragraph\n 4- get links that redirect to other Wikipedia pages\n")
    req = int(input("your request :"))
    while (req<0)or(req>5):
        print("nah, you can only stick to these 5 requests")
        print("1- see the content from Wikipedia\n 2- extract article title\n 3-Extract article text for each paragraph\n 4- get links that redirect to other Wikipedia pages\n")
        req = int(input("your request :"))
    
    if (req == 1):
        print("here is the content from the wikipedia page:")
        Get_Parse_Content(url)
    elif (req == 2):
        title = Extract_Title(url)
        print(f"the title of the article is {title}")
    elif (req == 3):
        content = extract_article_content_and_map(url)
        for heading, paragraph in content.items():
            print(f"{heading}: {paragraph}")
    elif (req == 4):
        links = Get_Links(url)
        for link in links:
            print(link)
            
wrapped_funtion()
        
```
